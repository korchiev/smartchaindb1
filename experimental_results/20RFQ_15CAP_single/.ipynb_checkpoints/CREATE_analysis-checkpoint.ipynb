```python
import pandas as pd
import numpy as np
import statistics as stat
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('scdb-20rfq-15cap-metrics.csv', names = ['Event', 'Time', 'Tx_type', 'Tx_id', 'Accept_id', 'time'])
```

```python
#grouping every transaction by its ID
grouped_data = data.groupby('Tx_id')
```

```python
tx_id = list(grouped_data.groups.keys())
len(tx_id)
```

```python
#new data frame to get the count of each transaction type
data_unique_ids = data.drop_duplicates(subset = ['Tx_id'])

create_count = 0
create_ids = []
for i in range(0, len(data_unique_ids)):
    if data_unique_ids.iloc[i]['Tx_type'] == 'CREATE':
        create_ids.append(data_unique_ids.iloc[i]["Tx_id"])
        create_count += 1   
print(f"#CREATEs =  {create_count}")
```

```python
df_1 = grouped_data.get_group(create_ids[5])
```

```python
df_1
```

```python
received_tx_pure_time = []
for i in range(0, len(create_ids)):
    df_1= grouped_data.get_group(create_ids[i])
    received_tx_pure_time.append(df_1[df_1['Event'] == 'received_tx']['Time'].to_list()[0])
```

```python
received_tx_pure_time
```

```python
#Calculating the time spend in each stage on Server
received_tx = []
validation_1 = []
validation_2_check_tx = []
validation_3_deliver_tx = []
end_block = []
commit_tx = []
for i in range(0, len(create_ids)):
    df = grouped_data.get_group(create_ids[i])
    received_time = df[df['Event'] == 'received_tx']['Time'].to_list()[0]
    received_tx.append(received_time)
    before_tendermint = df[df['Event'] == 'before_tendermint']['Time'].to_list()[0]
    validation_1.append(before_tendermint - received_time)
    check_time = df[df['Event'] == 'check_tx']['Time'].to_list()[0]
    validation_2_check_tx.append(check_time - before_tendermint)
    deliver_time = df[df['Event'] == 'deliver_tx']['Time'].to_list()[0]
    validation_3_deliver_tx.append(deliver_time - check_time )
    end_block_time =  df[df['Event'] == 'end_block']['Time'].to_list()[0]
    end_block.append(end_block_time - deliver_time)
    commit_time =  df[df['Event'] == 'commit_tx']['Time'].to_list()[0]
    commit_tx.append(commit_time - end_block_time)
```

```python
print(f" received_tx = {received_tx[0]}")
print(f" validation_1 = {validation_1[0]}")
print(f" validation_2_check_tx = {validation_2_check_tx[0]}")
print(f" validation_3_deliver_tx = {validation_3_deliver_tx[0]}")
print(f" end_block = {end_block[0]}")
print(f" commit_tx = {commit_tx[0]}")
```

```python
time_all_committed_creates = {}
time_passed_creates = []
create_time = {}
creates_overfit = []
creates_underfit = []
passed_validation = {}
other = []
for i in range(0, len(tx_id)):
    temp_list = []
    if "CREATE" == grouped_data.get_group(tx_id[i])["Tx_type"].to_list()[0]:
        temp_list = grouped_data.get_group(tx_id[i])["Time"].to_list()
        if len(temp_list) == 6 and grouped_data.get_group(tx_id[i])["Event"].to_list()[5] == "commit_tx":
            actual_time = max(temp_list)
            time_all_committed_creates[tx_id[i]] = actual_time
            time_passed_creates.append(actual_time)
            create_time[tx_id[i]] = actual_time
            if grouped_data.get_group(tx_id[i])["Event"].to_list()[1] == "before_tendermint":
                val = temp_list[1] - temp_list[0]
                passed_validation[tx_id[i]] = val            
        elif 2 <= len(temp_list) < 6 :
            if grouped_data.get_group(tx_id[i])["Event"].to_list()[1] == "before_tendermint" or grouped_data.get_group(tx_id[i])["Event"].to_list()[1] == "initial_validation_failed":
                val = temp_list[1] - temp_list[0]
                passed_validation[tx_id[i]] = val  
            temp_list.append(tx_id[i])
            creates_underfit.append(temp_list)
        elif 6 < len(temp_list) <= 10:
            temp_list = grouped_data.get_group(tx_id[i]).drop_duplicates(subset=['Event'])["Time"].to_list()
            val = temp_list[1] - temp_list[0]
            passed_validation[tx_id[i]] = val  
            actual_time = max(temp_list)
            time_passed_creates.append(actual_time)
            temp_list.append(tx_id[i])
            temp_list.append(actual_time)
            creates_overfit.append(temp_list)
            
        else:
            temp_list.append(tx_id[i])
            other.append(temp_list)
```

```python
print(f" #of unprocessed creates = {len(creates_underfit)}")
print(f" #of overprocessed creates = {len(creates_overfit)}")
print(f" #of processed creates = {len(time_passed_creates)}")
print(f" #of other creates = {len(other)}")
```
